```python
plt.figure(figsize=(6,5))
sns.countplot(x="gender",data=df,hue="status")
plt.title("Placed and Non-Placed Count By Gender")
plt.show()
```

​Description: A bivariate grouped bar chart analyzing the relationship between gender and placement success.
​Key Observations: 1. Out of the ~139 males, around 100 are placed (approx. 72% placement rate for men).
                   2. Out of the ~76 females, around 48 are placed (approx. 63% placement rate for women).

```python

# Count table
counts = pd.crosstab(df["gender"], df["status"])

# Convert to percentages within each gender
percent = counts.div(counts.sum(axis=1), axis=0) * 100

ax = percent.plot(kind="bar",
                  stacked=True,
                  figsize=(6,5),
                  color=["skyblue", "orange"])

# Add percentage labels
for container in ax.containers:
    labels = [f"{v:.1f}%" if v > 0 else "" for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="center", fontsize=10)

plt.title("Placement Percentage by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.ylim(0, 100)
plt.legend(title="Status")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns = [
    'ssc_p', 'ssc_b', 'hsc_p', 'hsc_b', 'hsc_s',
    'degree_p', 'degree_t', 'workex', 'etest_p',
    'specialisation', 'mba_p'
]

# Create subplots (4 rows × 3 columns = 12 spaces)
fig, axes = plt.subplots(4, 3, figsize=(18, 18))
axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i])

    axes[i].set_title(col, fontsize=12, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=45)

# Remove unused axes
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

​Conclusion: The count plots summarize the distribution of all major features in the dataset, highlighting the frequency of different categories and
             the spread of academic scores. This provides a clear understanding of the dataset's structure and helps identify patterns, imbalances,                 and dominant groups for further analysis.

```python
# Count table
counts = pd.crosstab(df["workex"], df["status"])

# Convert to percentages within each gender
percent = counts.div(counts.sum(axis=1), axis=0) * 100

ax = percent.plot(kind="bar",
                  stacked=True,
                  figsize=(6,5),
                  color=["skyblue", "orange"])

# Add percentage labels
for container in ax.containers:
    labels = [f"{v:.1f}%" if v > 0 else "" for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="center", fontsize=10)

plt.title("Placement Percentage by Work Experience")
plt.xlabel("Work Experience")
plt.ylabel("Percentage")
plt.ylim(0, 100)
plt.legend(title="Status")
plt.show()
```

# Students with work experience have a much higher placement rate (86.5%) compared to students without work experience (59.6%). This indicates that prior work experience improves placement opportunities.

## 📌 EDA Summary

### **Dataset Overview**

* **215** student records and **15** features.
* No duplicate values.
* Missing salary values (unplaced students) replaced with **0**.

### **Key Findings**

* **Placement Rate:** **68.84%** (148 students)
* **Not Placed:** **31.16%** (67 students)
* **Gender Distribution:** **64.65% Male**, **35.35% Female**
* **Most Common Specialization:** **Marketing & Finance (55.81%)**

### **Insights**

* Higher academic scores were generally associated with better placement.
* Students with **work experience** had a much higher placement rate (**86.5%**) than those without (**59.6%**).
* **Marketing & Finance** students earned higher average salaries than **Marketing & HR** students.
* Most placed students received salaries between **₹2–4 lakh**, with a few high-paying outliers.

### **Conclusion**

* Academic performance, work experience, employability skills, and specialization are the key factors influencing placement and salary.
* The dataset is clean and suitable for predictive modeling and further machine learning analysis.
